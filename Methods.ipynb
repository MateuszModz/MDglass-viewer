```python
def readFromFile(file,quantities):
    data=[]
    file_content=[]
    i,i1,i2=0,0,0
    for line in file:
        file_content.append(line)
        i+=1
        if 'Step' in line:
            i1=i
        if 'Loop' in line:
            i2=i
    quantities[0]=file_content[i1-1].split()
    for n in range(i1,i2-1):
        data.append([float(x) for x in file_content[n].split()])
    
    return data


```
